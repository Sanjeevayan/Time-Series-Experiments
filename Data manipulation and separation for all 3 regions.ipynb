```python
#### to filter region, closed category and fillna
def feature_engineering(df, region):
    df=df[df['Opportunity Geography'].isin(region)]
    df=df[df['Forecast Category'].isin(['Closed'])]
    df.loc[df['Closed Won Date'].isnull(), 'Closed Won Date']=df.loc[df['Closed Won Date'].isnull(),'Close Date']
    return df
```

```python
america_2017to18=feature_engineering(df_2011to18_1, ['Americas','SA-Americas'])
display(america_2017to18.sample(10))
```
